```python
item = 'Veiling'


meerdere_collecties = input("Gaat het om een veiling van meerdere collecties? ('ja','nee' of 'onbekend'): ")
if meerdere_collecties == 'ja':
    echte_eigenaar = ("Naam van de echte eigenaar (indien onbekend enter '-'): ")
    if echte_eigenaar == '-':
        namen_collecties = input("Geef namen van alle collecties indien niet meer dan vier (scheiden met een komma en een spatie', ')")
        not_section = input("Voor welke van de genoemde collecties is 'not section' van toepassing? (Indien geen enkele enter '-') : ")
        if not_section == '-':
            not_section = ''
    else:    
        eerste_verzamelaar = ("Naam eerste verzamelaar: ")+ ' et al.'
        
        
ter_plaatse_verkocht = input("Is het object ter plaatse verkocht? ('Ja', 'nee' of 'onbekend')")
if ter_plaatse_verkocht.lower() == 'ja':
    ter_plaatse_verkocht = 'sold on the premises'
else:
    ter_plaatse_verkocht = ''
        
        

if schakel.lower() == 'beginschakel' and item.lower() == 'veiling':
    inzenders_bekend = input("Zijn de inzenders bekend? (Enter: 'ja' of 'nee'): ")
    if inzenders_bekend.lower() == 'ja':
        start = 'Sale,'
        naam_eigenaar = input("Naam eigenaar: ") 
        geboortedatum = input("geboortedatum (dag-maand(voluit)-jaar) eigenaar(indien niet bekend enter '-'): ")
        sterfdatum = input("sterfdatum (dag-maand(voluit)-jaar) eigenaar(indien niet bekend enter '-'): ") 
        plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
        if sterfdatum == '-':
            sterfdatum = input("Blijkt uit de catalogus dat het om een postume veiling gaat?(enter 'ja' of 'nee'): ")
            if sterfdatum.lower() == 'ja':
                sterfdatum = '(†)'
            elif sterfdatum.lower() == 'nee':
                sterfdatum = ''
            else:
                print('Input ongeldig')
        if plaats_verzameling == '-':
            plaatsverzameling = ''
    elif inzenders_bekend.lower() == 'nee':
        start = 'Anonymous sale,'
        naam_eigenaar = ''
        geboortedatum= ''
        sterfdatum = ''
        plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
        if plaats_verzameling == '-':
            plaatsverzameling = ''
    else: 
        print('Input ongeldig')
        
if schakel.lower() in ['tussenschakel','eindschakel'] and item.lower() == 'veiling':
    inzenders_bekend = input("Zijn de inzenders bekend? (Antwoord 'ja', 'nee', ' reeds alleen initialen gegeven' of 'reeds exact gegeven'): ")
    if inzenders_bekend.lower() == 'ja':
        start = 'sale,'
        naam_eigenaar = input("Naam eigenaar: ")  
        geboortedatum = input("geboortedatum (dag-maand(voluit)-jaar) eigenaar (indien niet bekend enter '-'): ")
        sterfdatum = input("sterfdatum (dag-maand(voluit)-jaar) eigenaar(indien niet bekend enter '-'): ")
        plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
        if sterfdatum == '-':
            sterfdatum = input("Blijkt uit de catalogus dat het om een postume veiling gaat?(enter 'ja' of 'nee'): ")
            if sterfdatum.lower() == 'ja':
                sterfdatum = '(†)'
            elif sterfdatum.lower() == 'nee':
                sterfdatum = ''
            else:
                print('Input ongeldig')
        if plaats_verzameling == '-':
            plaatsverzameling = ''
    elif inzenders_bekend.lower() == 'nee':
        start = 'anonymous sale,'  
        naam_eigenaar = ''
        geboortedatum= ''
        sterfdatum = ''
        plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
        if plaats_verzameling == '-':
                plaatsverzameling = ''
    elif inzenders_bekend.lower() == 'alleen initialen':
        initialen = input("Naam voluit bekend? (Antwoord 'ja' of 'nee'): ")
        if initialen.lower() == 'ja':
            start = 'sale,'
            naam_eigenaar = input("Naam eigenaar: ")
            geboortedatum= ''
            sterfdatum = ''
            plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
            if plaats_verzameling == '-':
                plaatsverzameling = ''
        elif initialen.lower() =='nee':
            gender = input("Gender eigenaar(Antwoord 'm','f' of unknown): ")
            if gender == 'm':
                start = 'his sale,'
                geboortedatum= ''
                sterfdatum = ''
                plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
                if plaats_verzameling == '-':
                    plaatsverzameling = ''
            elif gender == 'f':
                start = 'her sale,'
                geboortedatum= ''
                sterfdatum = ''
                plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
                if plaats_verzameling == '-':
                    plaatsverzameling = ''
            elif gender == 'unknown':
                start = 'their sale,'
                geboortedatum= ''
                sterfdatum = ''
                plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
                if plaats_verzameling == '-':
                    plaatsverzameling = ''
            else: 
                print('input ongeldig')
        else:
            print('input ongeldig')
    elif inzenders_bekend.lower() == 'reeds exact gegeven':
        gender = input("Gender eigenaar(Antwoord 'm' of 'f')")
        naam_eigenaar = ''
        if gender == 'm':
            start = 'his sale,'
            geboortedatum = ''
            sterfdatum = ''
            plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
            if plaats_verzameling == '-':
                plaatsverzameling = ''
        elif gender == 'f':
            start = 'her sale,'
            geboortedatum: ''
            sterfdatum: ''
            plaats_verzameling = input("Plaats waar de verzameling zich bevond? (Alleen de plaatsnaam invoeren of '-'): ")
            if plaats_verzameling == '-':
                plaatsverzameling = ''
        else: 
            print('input ongeldig')
        
    else: 
        print('input ongeldig')
        
plaats_veiling = input("Plaats waar het item werd geveild? (Alleen de plaatsnaam invoeren of '-'): ")
if plaats_veiling == '-':
    plaats_veiling = ''
naam_veilinghuis = input("Naam veilinghuis(indien onbekend enter 'unknown'):")
if naam_veilinghuis == "Christies":
    naam_veilinghuis = "Christie's"
if naam_veilinghuis == "Sothebys":
    naam_veilinghuis = "Sotheby's"
if naam_veilinghuis.lower() == 'unknown':
    naam_veilingmeester = input("Naam veilingmeester(voorletters + achternaam)(indien onbekend enter '-'):")
    if naam_veilingmeester == '-':
        naam_veilinghuis = 'auction house not known'
    else:
        tweede_veilingmeester= 'and '+input("Naam tweede veilingmeester(voorletters + achternaam)(indien onbekend enter '-'): ")
        if tweede_veilingmeester == '-':
                naam_veilinghuis = naam_veilingmeester
        else:
            derde_veilingmeester = input("Indien nog meer veilingmeesters bekend enter 'ja' of 'nee': ")
            if derde_veilingmeester.lower() == 'ja':
                naam_veilinghuis = naam_veilingmeester+ tweede_veilingmeester,'et al.'
            if derde_veilingmeester.lower() == 'nee':
                naam_veilinghuis = naam_veilingmeester+ tweede_veilingmeester
    
dagen_veiling = input("Vond de veiling op meerdere dagen plaats? ('ja','nee' of 'unknown'): ")
if dagen_veiling.lower() == 'ja':
    begin_datum = input("Datum (dag-maand(voluit en begin met hoofdletter)-jaar) waarop de veiling zelf begon(Let op: gebruik geen spaties!): ")
    begin_datum_split = begin_datum.split('-')
    dag_verkocht= input("Datum (dag-maand(voluit en begin met hoofdletter)-jaar) waarop het object zelf werd geveild(Let op: gebruik geen spaties!): ")
    dag_verkocht_split = dag_verkocht.split('-')
    if begin_datum_split[1] != dag_verkocht_split[1]:
        datum_veiling = begin_datum_split[0]+' ('+dag_verkocht_split[0]+' '+dag_verkocht_split[1]+') '+ begin_datum_split[1]+' '+begin_datum_split[-1]+' ssq.'      
    elif begin_datum_split[-1] != dag_verkocht_split[-1]:
         datum_veiling = begin_datum_split[0]+' ('+dag_verkocht_split[0]+' '+ dag_verkocht_split[1]+ ' '+ dag_verkocht_split[-1]+') '+ begin_datum_split[1]+' '+begin_datum_split[-1]+' ssq.'
    else:
        datum_veiling = begin_datum_split[0]+' ('+dag_verkocht_split[0]+ ') '+ begin_datum_split[1]+' '+begin_datum_split[-1]+' ssq.'
        
elif dagen_veiling.lower() == 'nee' or 'unknown' :
    datum_veiling = input("Datum veiling (dag-maand(voluit)-jaar)(Indien onbekend enter '-'): ")
    if datum_veiling == '-':
        datum_veiling = ''
else:
    print('input ongeldig')

    
    
hors_catalogue = input("Gaat het om een verkoop buiten de catalogus ('ja' of 'nee'): ")
if hors_catalogue.lower() == 'ja':
    kavel = '(hors de catalogue)'
elif hors_catalogue.lower() == 'nee':
    kavel_bekend = input("Is het kavelnummer bekend? ('ja' of 'nee'): ")
    if kavel_bekend.lower() == 'ja':
        kavelnummer = input("Kavelnummer (alleen het getal): ")
        if kavelnummer.isdigit() == True:
            deel_kavel = input("Gaat het om een deel van een kavel('ja' of 'nee'): ")
            if deel_kavel.lower() == 'ja':
                x_number_kavel = input('Hoeveel andere objecten maken deel uit van dat kavel? (In cijfers!): ')
                if int(x_number_kavel) == '1':
                    kavel = 'no. '+kavelnummer+', with 1 other painting'
                elif int(x_number_kavel) > 1: 
                    kavel = 'no. '+kavelnummer+', with '+ x_number_kavel + ' other paintings'            
            elif deel_kavel.lower() == 'nee':
                kavel = 'no. '+kavelnummer
            else:
                print('input ongeldig')
        else:
            print('Alleen getallen invoeren!')
    elif kavel_bekend.lower() == 'nee':
        kavel = ''
    else: 
        print('input ongeldig')
else:
    print('input ongeldig')
```
